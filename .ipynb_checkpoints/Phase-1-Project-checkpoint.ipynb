```python
# API key
api_key = "cffa5c12fc10e5e1381fc438f7061c68"

# Fetching the genre data from theMovieDB API
genre_url = f"https://api.themoviedb.org/3/genre/movie/list?api_key={api_key}"
response = requests.get(genre_url)
genre_data = response.json()

# Creating a dictionary to map genre IDs to genre names
genre_dict = {genre['id']: genre['name'] for genre in genre_data['genres']}



# Function to map genre IDs to genre names
def map_genre_ids_to_names(genre_ids):
    genre_ids = ast.literal_eval(genre_ids)  # Convert string representation to actual list
    genre_names = [genre_dict.get(genre_id) for genre_id in genre_ids if genre_id in genre_dict]
    return genre_names

# Mapping the 'genre_ids' column to genre names
movies_data['genre_names'] = movies_data['genre_ids'].apply(map_genre_ids_to_names)

#Updated DataFrame
movies_data.head()
```

# Processing Movie Review Data

```python
#Merging the movie_info and reviews dataframes

merged_data_r = pd.merge(movie_info, reviews, on='id')
merged_data_r.head()
```

```python
merged_data_r = merged_data_r.drop(["synopsis", "rating_x", "dvd_date", "currency", "box_office", "fresh", "critic", "top_critic", "publisher", "date","director","writer"], axis=1)
merged_data_r.head()
```

```python
merged_data_r.info()
```

```python
# Dropping rows with no genre values
merged_data_r.dropna(subset=['genre'], inplace=True)

# Dropping the rows with NaN values in the "runtime" column
merged_data_r.dropna(subset=['runtime'], inplace=True)

# Converting the "runtime" column to numeric if it contains numeric values
merged_data_r['runtime'] = pd.to_numeric(merged_data_r['runtime'], errors='coerce')

# Dropping rows with null studio values
merged_data_r.dropna(subset=['studio'], inplace=True)

# Converting the 'rating_y' column to numeric if it contains numeric values
merged_data_r['rating_y'] = pd.to_numeric(merged_data_r['rating_y'], errors='coerce')

# Calculating the average rating
average_rating = merged_data_r['rating_y'].mean()

# Replacing missing values in 'rating_y' column with the average rating
merged_data_r['rating_y'].fillna(average_rating, inplace=True)

merged_data_r.head()

```

```python
# Converting theater_date column to datetime
merged_data_r['theater_date'] = pd.to_datetime(merged_data_r['theater_date'])
```


# 5. Data Analysis

# 1. Analyzing Box Office Performance

```python
#Analyzing the domestic and international gross earnings, budgets, and release dates

# Calculating the total gross earnings (domestic + worldwide)
merged_data_G["total_gross"] = merged_data_G["domestic_gross"] + merged_data_G["worldwide_gross"]

# Grouping the data by genre and calculating the average total gross for each genre
genre_performance = merged_data_G.explode("genre_names").groupby("genre_names")["total_gross"].mean().reset_index()

# Sorting the genres by average total gross in descending order
genre_performance = genre_performance.sort_values(by="total_gross", ascending=False)
```

```python
# Genres with the highest average total gross
top_performing_genres = genre_performance.head(10)
top_performing_genres.head(10)
```

```python
# Setting the plot style
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(x="total_gross", y="genre_names", data=top_performing_genres, orient="h")
plt.xlabel("Average Total Gross (in billions)")
plt.ylabel("Genre")
plt.title("Top Performing Genres at the Box Office")
plt.show()
```

- The genres with the highest total gross earnings are Animation and Adventure, with total gross earnings of 454.97 billion and 452.49 billion respectively. These genres demonstrate strong financial success and indicate a higher likelihood of performing well at the box office.

- Other genres that consistently perform well and have relatively high total gross earnings include Fantasy, Family, and Science Fiction, with total gross earnings of 417.07 billion, 387.22 billion, and 335.38 billion respectively.

- Genres such as Action, Comedy, War, Romance, and Crime also show solid performance at the box office, although their total gross earnings are relatively lower compared to Animation and Adventure.

- These results provide insights into the genres that have a higher likelihood of financial success in terms of box office performance. Microsoft can use this information to make informed decisions about genre selection and investment in film projects.







# 2. Assessing Genre Popularity and Trends

```python
movie_budgets['domestic_gross'] = movie_budgets['domestic_gross'].astype(float)
movie_budgets['worldwide_gross'] = movie_budgets['worldwide_gross'].astype(float)
movie_budgets['total_gross'] = movie_budgets['domestic_gross'] + movie_budgets['worldwide_gross']
movie_budgets['ROI'] = movie_budgets['total_gross'] / movie_budgets['production_budget']

movie_budgets['ROI'].head()
```

```python
top_10_movies = movie_budgets.sort_values('ROI', ascending=False).head(10)
```

```python
top_10_movies['movie']
```

```python
bottom_10_movies = movie_budgets.sort_values('ROI', ascending=True).head(10)
```

```python
bottom_10_movies['movie']
```

```python
# Converting 'release_date' column to datetime
movie_budgets['release_date'] = pd.to_datetime(movie_budgets['release_date'])

# Grouping the movies by year and count the number of successful movie releases per year
release_year_counts = movie_budgets[movie_budgets['ROI'] > 1].groupby(movie_budgets['release_date'].dt.year).size()

plt.figure(figsize=(10, 6))
plt.plot(release_year_counts.index, release_year_counts.values)
plt.xlabel('Year')
plt.ylabel('Number of Successful Releases')
plt.title('Number of Successful Movie Releases per Year')
plt.show()
```

- The number of successful movie releases has generally increased over the years, with some fluctuations. The industry has experienced growth and expansion over time.

- The 1960s and 1970s had a relatively lower number of successful releases compared to the subsequent decades. This may be due to various factors such as the evolving film industry, changing audience preferences, and technological advancements.

- The 1980s and 1990s witnessed a significant rise in the number of successful movie releases. This period marked a period of creativity, innovation, and blockbuster films.

- The late 1990s and early 2000s saw a substantial increase in the number of successful releases, reaching a peak in the early 2000s. This period was characterized by the emergence of franchises, sequels, and big-budget productions.

- From the mid-2000s onwards, the number of successful releases showed some fluctuations but remained relatively high. This suggests a consistent demand for movies despite market fluctuations and changing industry dynamics.

- The data indicates a decline in the number of successful movie releases in recent years (2017-2019). This could be influenced by various factors such as shifts in audience behavior, increased competition from other forms of entertainment, or changes in the film industry landscape

- The data highlights the dynamic nature of the movie industry, with periods of growth, peaks, and fluctuations. It showcases the evolving trends and preferences of moviegoers over time, reflecting the industry's ability to adapt and cater to audience demands.

```python
# Grouping the movies by month and count the number of successful movie releases per month
release_month_counts = movie_budgets[movie_budgets['ROI'] > 1].groupby(movie_budgets['release_date'].dt.month).size()

month_names = [calendar.month_abbr[i] for i in release_month_counts.index]

plt.figure(figsize=(10, 6))
plt.bar(month_names, release_month_counts.values)
plt.xlabel('Month')
plt.ylabel('Number of Successful Releases')
plt.title('Number of Successful Movie Releases per Month')
plt.show()
```

- The table shows the number of successful movie releases for each month. It indicates that December has the highest number of successful releases with 496, followed by October with 395. This suggests that these months may be popular for releasing successful movies.

- Seasonal Patterns: There appears to be a pattern of higher successful releases during certain months. The months of March, April, May, June, July, and August have relatively high numbers of successful releases, indicating a possible trend of releasing movies during the spring and summer seasons.

- These insights can help Micrososft in making informed decisions regarding release strategies, marketing campaigns, and resource allocation to maximize the chances of success for their films.

# 3. Assessing Genre Popularity and Trends

```python
# Sorting the movies data by popularity in descending order
movies_data_sorted = movies_data.sort_values(by='popularity', ascending=False)

# Flattening the lists of genre names into a single string
movies_data_sorted['genre_names'] = movies_data_sorted['genre_names'].apply(lambda x: x[0] if len(x) > 0 else "Unknown")

# Counting the occurrences of each genre
genre_counts = movies_data_sorted['genre_names'].value_counts()

# Removing the 'Unknown' genre from the counts
genre_counts = genre_counts[genre_counts.index != 'Unknown']

# Selecting the top 10 genres with the highest counts
top_10_genres = genre_counts.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_genres.values, y=top_10_genres.index, color='blue')
plt.title('Top 10 Genres with Highest Popularity')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()





```

- The top genre with the highest popularity is Drama

- The movie with the third-highest popularity is a combination of "Action, Adventure, Animation, Science Fiction" with a popularity score of 60.53.

- Genres such as "Adventure," "Science Fiction," and "Animation" appear in multiple entries, indicating their popularity among movie audiences.

- The presence of popular genres like "Action" and "Adventure" in the top entries suggests that these genres have a higher likelihood of attracting audience attention and generating popularity.

- The inclusion of genres like "Family," "Comedy," and "Fantasy" in some entries indicates the popularity of movies that cater to these genres' themes and target audiences.

 # 4. Evaluate Audience Ratings

```python
# Analyze the distribution of movie genres


# Sort the DataFrame by rating_y in descending order
top_10_genres = merged_data_r.sort_values(by='rating_y', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='genre', y='rating_y', data=top_10_genres)
plt.xticks(rotation=90)
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('Top 10 Genres with Highest Rating')
plt.show()



```

- "Animation|Comedy|Kids and Family" has a rating of 9.20, indicating it is also well-regarded.

- Genres such as "Drama," "Comedy," and "Action and Adventure|Mystery and Suspense" have a rating of 9.00, suggesting they are popular and well-received among viewers.





```python
#Exploring the Relationship between Runtime and Movie Rating

# scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(merged_data_r['runtime'], merged_data_r['rating_y'])
plt.title('Relationship between Movie Runtime and Audience Ratings')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Rating')
plt.show()
```

```python
# Selecting 'runtime' and 'rating_y' columns
runtime = merged_data_r['runtime']
rating = merged_data_r['rating_y']

# the correlation coefficient
correlation = np.corrcoef(runtime, rating)[0, 1]

print("Correlation coefficient between movie runtime and rating:", correlation)
```

- Based on the correlation coefficient, there is no meaningful relationship between movie runtime and rating. Other factors might have a more substantial influence on audience ratings.

```python

# Investigating the impact of studio on movie ratings
studio_ratings = merged_data_r.groupby('studio')['rating_y'].mean().dropna()
top_studios = studio_ratings.nlargest(10)
plt.figure(figsize=(10, 6))
top_studios.plot(kind='bar')
plt.title('Top 10 Studios with Highest Average Audience Rating')
plt.xlabel('Studio')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()


```

- This graph shows the average ratings for different movie studios. The Weinstein Company has the highest average rating of 9.00, followed by Lionsgate/Roadside Attractions with 8.90. This indicates that movies produced or distributed by these studios tend to receive higher ratings from viewers.

- Several independent studios are represented in the table, such as Roadside Attractions, Janus Films, Artisan, Destination Film, Entertainment One, Fine Line Features, First Independent Pictures, and First Look Pictures. Despite being independent, these studios have managed to achieve average ratings of 8.00 or higher. This highlights the potential for independent studios to produce high-quality movies that are well-received by viewers.

- The data emphasizes the significance of the studio in the success and perception of a movie. Different studios have different reputations and audience expectations, which can influence the ratings and overall reception of their movies. Studios with a strong track record of producing high-rated films may have a competitive advantage in the movie industry.

- These insights suggest that movie studios play a crucial role in shaping the perception and success of movies. Filmmakers and industry professionals can consider the reputation and performance of studios when making decisions about production, distribution, and partnerships. Additionally, viewers can use studio information as a reference point to discover movies that align with their preferences and expectations.
